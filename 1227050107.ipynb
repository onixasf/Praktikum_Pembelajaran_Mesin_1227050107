# Modul 2: Melakukan analisis data sederhana

1. Lakukan load data dengan menggunakan kode berikut:

```python
# %%
import pandas as pd

# %%
# reading the database
data = pd.read_csv("DatasetForCoffeeSales2.csv")

# %%
# printing the top 10 rows
display(data.head(10))
```

2. Cobalah lakukan filter data dengan melakukan penambahan kode seperti berikut:

```python
# %%
# Filter data only City, Category, Product, Quantity are showed

data.filter(["City", "Category", "Product", "Quantity"]).head()
```

3. Pada bagian tadi kita telah mencoba untuk melakukan filter data sehingga pada data yang ditampilkan hanya kolom City, Category, Product dan Quantity saja. Kemudian lakukan sorting dengan mencoba kode program berikut:

```python
# %%
# Sorting quantity (most)
data.sort_values("Quantity", axis=0, ascending=False, inplace=True, na_position="last")
display(data.head())
```

4. Setelah itu kita akan coba melakukan grouping untuk melihat total sales pada setiap City dengan melakukan grouping dengan mencoba kode program berikut:

```python
# %%
# Group Total Sales by City
grp1 = data.groupby('City')
result = grp1['Final Sales'].aggregate('sum')
print(result)
```

# Modul 3: Data Cleaning dan Transformation

1. Lakukan load data dengan menggunakan kode berikut:

```python
# %%
import pandas as pd

# %%
# reading the database
data = pd.read_csv("students.csv")

# %%
# printing the top 10 rows
display(data.head(10))
```

2. Tampilkan Informasi mengenai setiap fitur  yang kita miliki menggunakan fungsi info() lakukan seperti dibawah ini:

```python
# %%
# Show the dataset info
data.info()
```

3. Setelah itu perhatikan apakah terdapat Null value pada setiap fitur yang kita miliki dan apakah type data nya sudah sesuai untuk setiap kolom.
Berdasarkan hasil pengamatan terdapat beberapa data yang tidak lengkap (missing value) dalam fitur lunch, grade dan reading score, untuk lebih memastikannya gunakan fungsi isna() dan sum() pada dataset kita. Lakukan seperti dibawah ini: 

```python
# %%
# Missing Value Checking
data.isna().sum()
```

4. Untuk handling missing value tersebut, kita perlu memperhatikan terlebih dahulu type data dari setiap fitur nya, perhatikan fitur lunch, dikarenakan fitur lunch memiliki type data Object (String) dan hanya memiliki 2 jenis value (standard dan free/reduce) maka kita bisa handling missing value tersebut dengan nilai modus atau median nya, dalam kesempatan kali ini kita akan gunakan nilai modus nya. Lakukan seperti dibawah ini.

```python
# %%
# Filling Nan Value with mode
data['lunch'].fillna(data['lunch'].mode()[0], inplace=True)
```

5. Perhatikan fitur reading score, fitur ini memiliki type data float, maka kita dapat handling missing value nya dengan nilai mean. Lakukan seperti dibawah ini

```python
#%%
# Filling Nan Value with mean
data["reading score"].fillna(data["reading score"].mean())
```

6. Perhatikan fitur grade, fitur ini memiliki type data float, namun kita tidak dapat melakukan handling missing value dengan nilai mean dikarenakan value pada fitur ini merupakan data bertipe ordinal, maka kita bisa melakukan handling missing value dengan nilai modus dan median, kali ini kita akan gunakan nilai median. Lakukan Seperti dibawah ini.

```python
# %%
# Filling Nan Value with mode
data["grade"].fillna(data["grade"].median())
```
